```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_df=pd.read_excel('GDP Gas Brent.xlsm',sheet_name="gdp_raw")
df=data_df.loc[:,["Year","GDP"]] 
print(df)
```

```python
years = np.linspace(1900,2020,121)
sum_years = []
```

```python
for year in years:
    sum = 0
    for index, row in df.iterrows():
        if row["Year"] == year:
            sum+=row["GDP"]
    sum_years.append(sum)

print(sum_years)
```

```python
print(len(sum_years))
```

```python

exp = pd.DataFrame (sum_years, columns = ['GdpTot'])

exp.to_excel(r'GDPTot.xlsx', index = False)
```
